```python
# import library
import pandas as pd 
```

```python
# read the modified dataset
df = pd.read_csv('modified_dataset.csv')
df
```

```python
# filter dataset [and]
df.loc[  (df['Type 1'] == 'Grass' )  &  (df['Type 2'] == 'Fairy') ]
```

```python
# filter dataset [ | ] or
df.loc[ (df['Type 1'] == 'Grass' ) | (df['Type 2'] == 'Poison')]
```

```python
# filter using operator
df.loc[ df['Speed'] > 2]
```

```python
# filter using string parameter
df.loc[ df['Name'].str.contains('Mega')]
```

```python
# same as above but using not 
df.loc[  ~df['Name'].str.contains('Mega') ]
```

```python
# aggreate statistic

# all pokemon with highest attack values
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
# all pokemons with lowest attack values
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=True)
```

```python
# all pokemons with higest defence values
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
# count number of types of dargon
df.groupby(['Type 1', 'Type 2']).count()
```
